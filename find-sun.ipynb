```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
import ipywidgets as widgets
from IPython.display import clear_output


def get_image(filename='../input/sun-image/215221e8a54907ec2788c7d39c17748d (2).jpg'):
    image = cv2.imread(filename)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return image

def plot_image(image):
    fig, ax = plt.subplots(1, 1, figsize=(12, 7))
    ax.imshow(image, cmap='gray')
    plt.show()
```

```python
original_image = get_image()
plot_image(original_image)
```

```python
image = get_image()
kernel_size = int(image.shape[1] / 8)
current_values = [ 0 for i in range(8) ]
kernel_element_count = kernel_size * kernel_size

result = []

height, width = image.shape 

index_offset = 0

while True:    
    for i in range(index_offset, min(index_offset + kernel_size, height)):
        for j in range(width):
            current_values[j // kernel_size] += image[i, j]
    
    threshold_value = 127
    
    current_values = map(lambda x: int(x / kernel_element_count), current_values)
    current_values = list(map(lambda x: 1 if x > threshold_value else 0, current_values))
    
    byte = 0
    for value, i in zip(reversed(current_values), range(len(current_values))):
        byte += value * pow(2, i)
    
    result.append(byte)
    index_offset += kernel_size
    current_values = [ 0 for i in range(8) ]
    
    if index_offset >= height:
        break
    

result
```

```python
def bitfield(n):
    binary = bin(n)[2:]
    binary = f'{"0" * (8 - len(binary))}{binary}'
    
    return [int(digit) for digit in binary]


back_image = map(bitfield, result)
back_image = map(lambda x: list(map(lambda el: el * 255, x)), back_image)
back_image = np.array(list(back_image))
```

```python
plot_image(back_image)
```
